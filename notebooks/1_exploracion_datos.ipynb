```python
# Identificaci√≥n de valores at√≠picos usando el m√©todo IQR
def deteccion_outliers_iqr(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    outleirs = data[(data < q1 - 1.5 * iqr) | (data > q3 + 1.5 * iqr)]
```

```python
outliers_temp = deteccion_outliers_iqr(df_troll_a_2024['TEMP'])
print("\nValores at√≠picos en la columnas 'TEMP': ")
print(outliers_temp)
```

```python
outliers_DEWP = deteccion_outliers_iqr(df_troll_a_2024['DEWP'])
print("\nValores at√≠picos en la columnas 'DEWP': ")
print(outliers_DEWP)
```

```python
outliers_SLP = deteccion_outliers_iqr(df_troll_a_2024['SLP'])
print("\nValores at√≠picos en la columnas 'SLP': ")
print(outliers_SLP)
```
